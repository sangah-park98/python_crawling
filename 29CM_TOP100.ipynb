```python
import warnings
warnings.filterwarnings(action='ignore')
%config Computer.use_jedi = False
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
```

```python
# TOP100 패션 브랜드 크롤링
def getName():
    targetSite = 'https://shop.29cm.co.kr/best-items?category_large_code=268100100'
    request = requests.get(targetSite)
    html = request.text
    soup = BeautifulSoup(html, 'html.parser')
    
    brandnames = []
    divs = soup.findAll('div',{'class': 'css-1ed1w2n e1l5w7nb4'})
    for div in divs:
        chapter = div.findAll('a')
        for a in chapter:
            brandnames.append(a.text)
    return brandnames
```

```python
brandnames = getName()
for i in range(len(brandnames)):
    print('{:2d}. {}'.format(i, brandnames[i]))
```
