```python
import warnings
warnings.filterwarnings(action='ignore')
%config Computer.use_jedi = False
from bs4 import BeautifulSoup
```

```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome('./chromedriver.exe')
driver.get('https://shop.29cm.co.kr/best-items?category_large_code=268100100')
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
brandnames = soup.findAll('a', {'class': 'css-1irnxxf e1l5w7nb5'})
prices = soup.findAll('div', {'class': 'css-3jywan e1l5w7nb6'})


for i in range(len(brandnames)):
    brandname = brandnames[i].text.strip()
    price_text = prices[i].text.strip()
    price = price_text.split('%')[1] if '%' in price_text else price_text
    print('{:2d}위: {}, 가격: {}원'.format(i + 1, brandname, price))
```
