```python
import warnings
warnings.filterwarnings(action='ignore')
%config Computer.use_jedi = False
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
```

```python
targetSite = 'https://www.wconcept.co.kr/topseller'
header = {'User-agent': 'Mozilla/5.0 (Windows NT 6.3; WOW64; Trident/7.0; rv:11.0) like Gecko'}
request = requests.get(targetSite, headers=header)
print(request)
```

```python
html = request.text
# print(html)
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
brandnames = soup.findAll('div', {'class', 'brand'})
# print(brandnames)
for i in range(len(brandnames)):
    print('{:3d}위 {}'.format(i+1, brandnames[i].text.strip()))
    
```
