```python
import warnings
warnings.filterwarnings(action='ignore')
%config Computer.use_jedi = False
from bs4 import BeautifulSoup
```

```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome('./chromedriver.exe')
driver.get('https://shop.29cm.co.kr/best-items?category_large_code=268100100')
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
brandnames = soup.findAll('a', {'class': 'css-1irnxxf e1l5w7nb5'})
for i in range(len(brandnames)):
    print('{:3d}위 {}'.format(i+1, brandnames[i].text.strip()))
```
